Desde langchain se acumulan los mensajes

```python
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage, ToolMessage

ai_msg = AIMessage(content="Message from the AI")
ai_msg.text
```

```python
human_msg = HumanMessage(content="Message from the human")
human_msg.text
```

```python
history = [ai_msg, human_msg]
history
```

```python
for msg in history:
    msg.pretty_print()
```

Para seguir añadiendo mensajes al historial, se debe seguir la convención de concatenación de listas

```python
history = history + [AIMessage(content="Another message from the AI")]
for msg in history:
    msg.pretty_print()
```
