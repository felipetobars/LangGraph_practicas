#### Uso de plantillas para los Prompts

```python
from langchain_core.prompts import PromptTemplate
```

```python
# El \ al inicio del prompt es para indicar que el string continúa en la siguiente línea, lo que mejora la legibilidad del código () ahorra una linea vacía que consume tokens
template = """\
# Instrucciones para el Asistente de IA

Recuerda que hoy es {fecha_actual}

## Rol
Eres un experto en marketing digital con 10 años e experiencia

## Tarea
Analiza el siguiente texto de un anuncio y proporciona 3 sugerencias para mejorarlo, enfocándote en el llamado a acción

## Texto del Anuncio
{texto_anuncio}

## Formato de Salida
- *Sugerencia 1:* [Tu primera sugerencia]
- *Sugerencia 2:* [Tu segunda sugerencia]
- *Sugerencia 3:* [Tu tercera sugerencia]
"""

prompt_template = PromptTemplate.from_template(template)
prompt = prompt_template.format(fecha_actual="2024-06-01", texto_anuncio="Compra nuestro producto ahora y obtén un descuento del 20%")
print(prompt)
```

#### Si no se envía nada en format() daraá error, por lo que se pueden definir valores por defecto en la plantilla, por ejemplo:

```python
import datetime
today = datetime.datetime.now().strftime("%Y-%m-%d")
prompt_template = PromptTemplate.from_template(template, partial_variables={"fecha_actual": today})
prompt = prompt_template.format(texto_anuncio="Compra nuestro producto ahora y obtén un descuento del 20%")
print(prompt)
```

#### Usando Jinja2 para plantillas más complejas, por ejemplo con lógica de control de flujo:

```python
# El - en el {% endif -%} es para eliminar el salto de línea que normalmente se genera al finalizar un bloque de control, lo que ayuda a mantener el formato del prompt más limpio.
template = """\
    You are a helpful assistant that responds to the user's message

    {%if name %}
    The user's name is {{name}} and you can call them by that name
    {% endif -%}

    The user's message is: {{user_message}}
    """
prompt_template = PromptTemplate.from_template(template, template_format="jinja2")
prompt = prompt_template.format(user_message="Cómo puedo optimizar mi sitio web?")
print(prompt)
```
