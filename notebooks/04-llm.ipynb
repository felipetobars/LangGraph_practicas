Incorporando LLMs a LangGraph

```python
from langchain_ollama import ChatOllama

qwen25_llm = ChatOllama(model="qwen2.5:3b", temperature=0)
response = qwen25_llm.invoke("Hola cómo estás?")
response.pretty_print()
```

```python
gemma3_llm = ChatOllama(model="gemma3:4b", temperature=0)
response = gemma3_llm.invoke("Hola cómo estás?")
response.pretty_print()
```

Ver como si fuera un típico historial de chat

```python
from langchain_core.messages import AIMessage, HumanMessage, SystemMessage

msg_1 = SystemMessage(content="You are a helpful assistant that talks in spanish")
msg_2 = HumanMessage(content="Me llamo Luis")
msg_3 = AIMessage(content="Hola Luis, cómo estas?")
msg_4 = HumanMessage(content="Cómo me llamo?")

history = [msg_1, msg_2, msg_3, msg_4]
response = gemma3_llm.invoke(history)
response.pretty_print()
```
